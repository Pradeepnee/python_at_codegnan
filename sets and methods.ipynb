### Set Methods
- Adding Elements To Set
   - add(ele): it adds element to set if element not exists in set, otherwise nothing
   - update(iterable): It adds all iterable elements to set
- Removing elements from set
   - remove(ele): It removes the element from the set if element exists in set otherwise 'KeyError'
   - discard(ele): It removes the element from the set if element exists in set otherwise nothing
   - pop(): It removes and returns the elements from set 
- Set Operations
   - union(another_sets): It returns the all common elements from the all sets
   - intersection(another_set): It returns the common elements from the all sets
   - difference(another_set): It returns the elements those are not lies in another_sets
   - symmetric_difference(another_set): It returns the Non-common elements from all sets

```python
s1={10,20,30,40,50}
s2={30,40,60,70,90,10}
s1.add(90)
print(s1)
s2.remove(30)
print(s2)
s2.add(30)
s3=s1.union(s2)      # '|' for union
print(s3)
s4=s1.intersection(s2)  # '&'    for intersection
print(s4)
s5=s2.difference(s1)     #'-' for difference
print(s5)
s6=s1.difference(s2)
print(s6)
s7=s1.symmetric_difference(s2)   # '^' for Symmetric Difference
print(s7)
# s1.remove(1)   # Key Error
s9=s1.discard(1)
print(s9)
```
