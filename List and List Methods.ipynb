# List Methods
  - Element Adding Methods
    - append(item): It adds at the end of the list (adds one at a time)
    - extend(iterable) : it adds all iterable elements at the end of the list(adds multiple elements at a time)
    - insert(pos, item) : This Method adds the item at specified position

```python
lst=[1,2,3,4]
print(lst.append(10))
print(lst)
lst.append("code")
print(lst)
lst.append([10,20,30])
print(lst)
print(lst[6])
lst.extend([50,60,70])
print(lst)
lst.insert(1,20000)
print(lst)
```

- Element Removing Methods
  - remove(val): It removes first occurance of the value if value exists in list , otherwise it raises value error.
  - pop(): It remove and returns the last element of list.
  - pop(pos): It remove and returns the specified Position element from the list.
  - clear():It removes all the elements from the list.
  - del(): It is python keyword ,used to delete values.

```python
lst=[10,20,30,40,20]
lst.remove(20)
print(lst)
#lst.remove(100)  - ValueError: 100 not in lst
val=lst.pop()
print(val)
print(lst)
val2=lst.pop(1) #  it removes the 1st position element
print(val2,lst)
# print(lst.pop(10)) - IndexError: pop index out of range
print(lst)
del lst[0]
print(lst) # deletes 10 i.e(index 0)
lst.clear()
print(lst)
```

```python
# index(val): It returns the index of first occurence of value if value exists in list Otherwise ValueError
# count(val): It returns the total frequenc of value
lst=[10,20,30,40,50,50,'shamp']
print(lst.index(50))
print(lst.count(50))
print(lst.count(100))
print(lst.count(0))
```

```python
# Sort (reverse=false): It sorts the all list elements in ascending order, If reverse value is True it sorts in Descending Order.
lst=[5,43,63,6,73,-2,4,-6]
lst.sort()
print(lst)
print(lst.sort(reverse=True))
print(lst)
```
