# Built in Functions : The Functions defination is provided by python , those functions called as Built in Functions


```python
#abs(): It returns the absolute value
print(abs(20-10))
print(abs(10-20))
```

```python
# round(val,pos): This function rounds the float value to particular positions
print(round(2.2589,2))
print(round(2.2539,2))
print(round(2.2539,3))
print(int(2.2539))
```

```python
# ord(char): It returns the ASCII value of character
# chr(ascii_val): It returns the character for specified ASCII value
print(ord("A"))
print(chr(65))
```

```python
# sorted(iterable, reverse=False, Key=function) : It returns new sorted iterable
# sort the fruits based on the length of fruit name
fruits=['Apple','banana','kivi','Mango','water Melon']
res=sorted(fruits)
print(res)
print(sorted(fruits,reverse=True))
print(sorted(fruits,reverse=True,key=len))  # It sorts the list based on the length of each item or element length 
```

```python
# map(function, iterable)
# find the square of all list of integers using map function
def square(x):
    return x**2
arr=[1,2,3,4,5]
res=list(map(square,arr))
print(res)
# filter (function,iterable)
```

```python
#filter(function, iterable)
# find even numbers from list of numbers
def check_even(x):
    if x%2==0:
        return x
lst=[1,2,3,4,5,6]
res=list(filter(check_even,lst))
print(res)
```

```python
# map(function, iterable)
# find even numbers from list of numbers
def check_even(x):
    if x%2==0:
        return x
lst=[1,2,3,4,5,6]
res=list(map(check_even,lst))
print(res)
```

```python
# find the 3's multipliers from list of integers
def mut(x):
    if x%3==0:
        return x
lst=[1,24,54,22,44,51,33,15]
res=list(filter(mut,lst))
print(res)
```

```python
# enumerate(iterable): This function returns the list of tuple of index and item from iterable
fruits=['Mango','Apple','banana','kivi','Water Melon']
print(enumerate(fruits))
print(list(enumerate(fruits))) 
for ind,val in enumerate(fruits):
    print(f"{ind}-{val} and length is {len(val)}")
```

```python
# tuple unpacing
t=(2,3)
a,b=t
print(a,b)
```

```python
# zip(iterable1, iterable2,.......): It returns all iterables as single iterable of tuples
roll_no=[101,102,103,104,105]
names=['Ravi','Raju','siva','Rani','Gopi']
percentage=[90,67,96.8,67.65,99.99999]
print(list(zip(roll_no,names,percentage)))
```

```python
# num conversion
num=int(input())
res=""
while num>0:
    bit=num%2
    num=num//2
    res=str(bit)+res
print(res)
```

```python
# any(): This Function returns True when any one iterable items return True
# all() : This Function returns True when all iterable iems return True
arr=[1,2,3,4,0]
print(all(arr))
print(any(arr))
```

```python
percentage=[34,35,67,32]
if all(percentage)>35:
    print("All students passed")
else:
    print("all students Not Passed")
```

```python
percentage=[45,64,23,98,56]
for num in percentage:
    if num<35:
        print("All Students not Passed")
        break
else:
    print("All Students pass")
```

```python
percentage=[45,64,23,98,56]
lst=[]
for num in percentage:
    if num<35:
        lst.append(False)
    else:
        lst.append(True)
print(lst)
if all(lst):
    print("ALL passed")
else:
    print("All Students not passed")
```
