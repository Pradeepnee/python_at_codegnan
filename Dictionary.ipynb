### Dictionary Methods
   - Get Value
   - get(key,default_val):It returns the value if key exists in dictionary otherwise default_val
   - update value and Add items to Dictionary
        -update(new_dictionary): Dictionary Updates with new_dictionary Items
   - Remove Items from dictionary
       -pop(key): It Returns value and remove item if key is exists in dictionary otherwise 'KeyError'
       - popitem(): It returns tuple of key value pair and removes the last item from dictionary
   - Get Keys, value and Items from dictionary
       - keys(): It returns list of dictionary values
       - values(): it returns list of dictionary values
       - items(): It returns list of tuple key-value pairs

```python
details={'name':'Ravi','age':20,'course':'python','marks':[85,90,86],'age':50}
 #print(details['location'])  #key error because location is not there 
print(details.get('name'))
print(details.get('location'))
print(details.get('age',18))
print(details.get('location',"Location key not exists"))
```

```python
details={'name':'Ravi','age':20,'course':'python'}
d={'marks':[85,90,86],'age':50}
print(details.update(d))
print(details)
```

```python
details={'name':'Ravi','age':20,'course':'python','marks':[85,90,86],'age':50}
print(details.pop('age'))
print(details)
print(details.popitem())     #pop item return type is tuple
print(details)
 # print(details.pop("percentage"))     - key error
print(details.keys())
d={'location':'hyd'}
print(details.update(d))
print(details)
print(details.values())
print(details.items())
```
