### String Methods
- case conversion methods
  - upper():it converts entire string to uppercase
  - lower(): it converst entire string into lowercase
  - title(): it converts every word first character int o uppercase and remaining into lower
  - swapcase():it converts lower into upper and upper to lower
  - capitalize(): it converts the first character of entire string into uppercase and remaining into lower case

```python
string="MY name is hence"
var=string.upper()
print(var)
```

```python
string="MY name is HENCE"
var=string.lower()
print(var)
```

```python
string="MY name is HENCE"
var=string.title()
print(var)
```

```python
string="MY na34455me is HENCE"
print(string.swapcase())
print(string.capitalize())
print(string.upper())
print(string.lower())
print(string.title())
```

```python
# White Space Remove methods
# strip: it removes the white space to both sides of string
# Lstrip():it removes the  white space which is left side of the string
# rstrip():it removes the white space which is right side of the string
string= "       Codegnan         "
print(string)
print(string.strip())
print(string.lstrip())
print(string.rstrip())
```

```python
# White Space Remove methods
# strip: it removes the white space to both sides of string
# Lstrip():it removes the  white space which is left side of the string
# rstrip():it removes the white space which is right side of the string
string= "@@@@@@Code    gnan@@@@@@@"
print(string)
print(string.strip("@"))
print(string.lstrip("@"))
print(string.rstrip("@"))
```

```python
#split(): returns the list of substring and by default splits at white space
string="The Ocean is blue and skyis Blue 23456@"
print(string.split())
print(string.split("is"))
print(string.split("Z"))
print(string.split("blue"))
print(string.split("@"))#returns empty string if it is in end of string
```

```python
# Indexing finding methods
# index(sub_string) it returns the first occurence index of sub_string string if exists otherwise "ValueError"
# find(sub_string) it returns the first occurence index of sub_string string if exists otherwise -1
string="The ocean is bluue and sky is blue"
print(string.index("o"))
print(string.index("is"))
print(string.index("is",14))# it starts checking from 14th index
```

```python
# Indexing finding methods
# index(sub_string) it returns the first occurence index of sub_string string if exists otherwise "ValueError-sub_string not found"
# find(sub_string) it returns the first occurence index of sub_string string if exists otherwise -1
string="The ocean is bluue and sky is blue"
print(string.find("o"))
print(string.find("is"))
print(string.find("is",14))# it starts checking from 14th index
print(string.find("z"))
```

```python
# replace method
# replace(old, new): it replaces old_string with new_string in string
string="The ocean is blue and sky is blue"
print(string.replace('blue','red'))
print(string.replace('blue','red',1))# replaces only in 1 occurance
print(string.replace('blue','red',0))
```

```python
# "sep".join(iterable): itt joins the all iterable elements with seperator
lst=['The', 'Ocean', 'is', 'blue', 'and', 'skyis', 'Blue', '23456@']
print("".join(lst))
print(" ".join(lst))
```

### string checking methods
    - islower(): returns true when entire string is in lowercase
    - isupper(): returns true when entire string is in uppercase
    - istitle():returns true when entire string is in title format
    - isalpha(): returns true when entire string contains alphabets
    - isalnum(): returns true when entire string contains alphabets or digits
    - isdigit(): returns true when entire string contains digits
    - startswith(sub_string): returns true when string starts with the  sub_string
    - endswith(sunstring): returns true when string ends with sub_string 





```python
string="TheOceanisblueue2andSkisBlu1234e"
print(string.islower())
print(string.isupper())
print(string.istitle())
print(string.isalpha())
print(string.isalnum())
print(string.startswith("T"))
print(string.endswith("e"))
```
