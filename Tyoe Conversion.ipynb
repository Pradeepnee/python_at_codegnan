```python
# Reading integer from user
# integer to float, string , bool
n=int(input())
print(type(n),n)
print(float(n))
print(str(n))
print(bool(n))
```

```python
# reading float value from user
# float to int, string, bool
val=float(input())
print(type(val),val)
print(float(val))
print(str(val) [:2])
print(bool(val))
print(int(val))
```

```python
# reading complex value
real=int(input())
img=int(input())
c=complex(real,img)
print(type(c),c)
```

```python
# reading string value from user
s=input()  # '123'
print(type(s),s)
print(int(s))
print(float(s))
print(list(s))    # it returns list of characters
print(tuple(s))   # it returns tuple of characters
print(set(s))     # it returns set of characters
```

```python
s=input()  # '123'
print(type(s),s)
l=list(s)
print(l[2:])    # it returns list of characters
print(tuple(s))   # it returns tuple of characters
print(set(s))
```

```python
# list to tuple,set, string
l=list(map(int,input().split()))
print(l)
print(tuple(l))
print(set(l))
s=str(l)
print(s[1])
```

```python
# read list of strings
t=list(map(str,input().split()))
print(list(t))
print(set(t))
print(str(t))
```
